```python
notes = []

for file in glob.glob("midi_songs/*.mid"):
    midi = converter.parse(file) # Convert file into stream.Score Object
    
    print("parsing %s"%file)
    
    elements_to_parse = midi.flat.notes
    
    
    for ele in elements_to_parse:
        # If the element is a Note,  then store it's pitch
        if isinstance(ele, note.Note):
            notes.append(str(ele.pitch))

        # If the element is a Chord, split each note of chord and join them with +
        elif isinstance(ele, chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
```

```python
with open('notes','rb') as f:
    notes = pickle.load(f)
```

```python
n_vocab=len(set(notes))
```

```python
n_vocab 
```

```python
seq_length=100
```

```python
pitchnames = sorted(set(notes))
```

```python
ele_to_int = dict( (ele,num) for num,ele in enumerate(pitchnames))
int_to_ele = dict( (num,ele) for num,ele in enumerate(pitchnames))
```

```python
network_input = []
network_output = []
```

```python
for i in range(len(notes)-seq_length):
    seq_in = notes[i:i+seq_length]
    seq_out = notes[i+seq_length]
    
    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
n_patterns = len(network_input)
print(n_patterns)
```

```python
network_input = np.reshape(network_input,(n_patterns,seq_length,1))
```

```python
network_input.shape
```

```python
normalised_network_input = network_input/float(n_vocab)
```

```python
normalised_network_input
```

```python
network_output = np_utils.to_categorical(network_output)
```

```python
network_output.shape
```

```python
from keras.models import Sequential,load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint,EarlyStopping


```

```python
model = Sequential()
model.add(LSTM(units=512,input_shape=(network_input.shape[1],network_input.shape[2]),return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam')
```

```python
model.summary()
```

```python
checkpoint = ModelCheckpoint('model.hdf5',monitor='loss',save_best_only=True,mode='min')
model_his = model.fit(normalised_network_input,network_output,epochs=100,batch_size=64,callbacks=[checkpoint])
```

```python
model = load_model('model.hdf5')
```

```python
network_in=[]
for i in range(len(notes)-seq_length):
    seq_in = notes[i:i+seq_length]    
    network_in.append([ele_to_int[ch] for ch in seq_in])
   
```

```python
start = np.random.randint(len(network_in)-1)
pattern = network_in[start]
prediction_out = []

for note_index in range(200):
    prediction_in = np.reshape(pattern,(1,len(pattern),1))
    prediction_in = prediction_in/float(n_vocab)
    prediction = model.predict(prediction_in)
    
    idx = np.argmax(prediction)
    result = int_to_ele[idx]
    prediction_out.append(result)
    
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
prediction_out
```

```python
offset = 0
output_notes = []

for pattern in prediction_out:
    
    if('+' in pattern) or pattern.isdigit():
        notes_in_chord = pattern.split('+')
        temp_notes=[]
        for current_node in notes_in_chord:
            new_note = note.Note(int(current_node))
            new_note.storedInstrument = instrument.Piano()
            temp_notes.append(new_note)
            
        new_chord = chord.Chord(temp_notes)
        new_chord.offset = offset
        output_notes.append(new_chord)
        
    else:
        new_note = note.Note(pattern)
        new_note.offset = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
        
    offset += 0.5
        
        
        
       
    
```

```python
output_notes
```

```python
midi_stream = stream.Stream(output_notes)
midi_stream.write('midi',fp='./generated_music/music1.mid')
```
